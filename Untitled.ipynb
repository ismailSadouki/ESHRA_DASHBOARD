```python
import pandas as pd

# Load the data
```

```python
df = pd.read_csv("building_consumption.csv")
```

```python
# Ensure proper datetime format
df['timestamp'] = pd.to_datetime(df['timestamp'])
df.head()
```

```python
total_per_campus = df.groupby(['timestamp', 'campus_id'])['consumption'].sum().reset_index()

import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
for campus in total_per_campus['campus_id'].unique():
    data = total_per_campus[total_per_campus['campus_id'] == campus]
    plt.plot(data['timestamp'], data['consumption'], label=f'Campus {campus}')

plt.legend()
plt.title('Total Consumption per Campus Over Time')
plt.xlabel('Date')
plt.ylabel('Consumption')
plt.tight_layout()
plt.show()
```

```python
df['month'] = df['timestamp'].dt.to_period('M')
monthly_trend = df.groupby('month')['consumption'].sum().reset_index()
monthly_trend['month'] = monthly_trend['month'].dt.to_timestamp()

plt.figure(figsize=(10,5))
plt.plot(monthly_trend['month'], monthly_trend['consumption'])
plt.title('Monthly Total Consumption')
plt.xlabel('Month')
plt.ylabel('Consumption')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
avg_per_meter = df.groupby('meter_id')['consumption'].mean().sort_values(ascending=False)
print(avg_per_meter.head())
```

```python
import seaborn as sns

df['hour'] = df['timestamp'].dt.hour
df['day_of_week'] = df['timestamp'].dt.day_name()

heatmap_data = df.groupby(['day_of_week', 'hour'])['consumption'].mean().unstack()

# Reorder days
ordered_days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
heatmap_data = heatmap_data.reindex(ordered_days)

plt.figure(figsize=(12,6))
sns.heatmap(heatmap_data, cmap='YlGnBu')
plt.title('Average Consumption by Hour and Day of Week')
plt.xlabel('Hour of Day')
plt.ylabel('Day of Week')
plt.tight_layout()
plt.show()
```

```python
# Total consumption by date
daily = df.groupby(df['timestamp'].dt.date)['consumption'].sum()

daily.plot(figsize=(12,5), title='Daily Total Consumption')
plt.ylabel('Consumption')
plt.xlabel('Date')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
campus_total = df.groupby('campus_id')['consumption'].sum()

campus_total.plot(kind='bar', title='Total Consumption by Campus', ylabel='Consumption', xlabel='Campus', figsize=(8,5))
plt.tight_layout()
plt.show()
```

```python
df['month'] = df['timestamp'].dt.to_period('M')

monthly = df.groupby(['month', 'campus_id'])['consumption'].sum().reset_index()
monthly['month'] = monthly['month'].dt.to_timestamp()

plt.figure(figsize=(12,6))
for campus in monthly['campus_id'].unique():
    campus_data = monthly[monthly['campus_id'] == campus]
    plt.plot(campus_data['month'], campus_data['consumption'], label=f'Campus {campus}')
    
plt.legend()
plt.title('Monthly Consumption per Campus')
plt.xlabel('Month')
plt.ylabel('Consumption')
plt.tight_layout()
plt.show()
```

```python
df['hour'] = df['timestamp'].dt.hour
df['weekday'] = df['timestamp'].dt.day_name()

heatmap = df.groupby(['weekday', 'hour'])['consumption'].mean().unstack()
heatmap = heatmap.reindex(['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'])

plt.figure(figsize=(12,6))
sns.heatmap(heatmap, cmap='viridis')
plt.title('Average Hourly Consumption (by Day)')
plt.xlabel('Hour')
plt.ylabel('Weekday')
plt.tight_layout()
plt.show()
```

```python
recent = df[df['timestamp'] >= df['timestamp'].max() - pd.Timedelta(days=7)]
daily_recent = recent.groupby(recent['timestamp'].dt.date)['consumption'].sum()

daily_recent.plot(kind='bar', figsize=(10,5), title='Consumption Over Last 7 Days')
plt.ylabel('Consumption')
plt.xlabel('Date')
plt.tight_layout()
plt.show()
```

```python
top_meters = df.groupby('meter_id')['consumption'].sum().sort_values(ascending=False).head(10)

top_meters.plot(kind='barh', figsize=(8,6), title='Top 10 Highest Consuming Meters')
plt.xlabel('Total Consumption')
plt.tight_layout()
plt.show()
```
